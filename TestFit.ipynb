```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
xdata = [1, 2, 3, 4, 5]
xdata = np.asarray(xdata)
ydata = np.exp(xdata)
ydata = np.asarray(ydata)

print(xdata, ydata)
#np.exp(x)
```

```python
plt.scatter(xdata, ydata, color="red", marker=".")
```

```python
def Func_to_Fit(x, A, B):
    y = A*np.exp(B*(x))
    return y

guess = [1.5, 1]
parameters, covariance = curve_fit(Func_to_Fit, xdata, ydata, p0 = guess)
print(parameters)

x2 = np.arange(0, 5, 0.01)

y2 = Func_to_Fit(parameters[0], parameters[1], x2)

plt.scatter(xdata, ydata)
plt.plot(x2, y2)

```
